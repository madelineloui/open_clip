```python
import os
import glob
import pandas as pd
import csv
from tqdm import tqdm
```

### Create training csv with img path, caption

```python
txt_dir = '/home/gridsan/manderson/vlm4rs/fmow/fmow-gpt/train/landmarks-only'
img_dir = '/home/gridsan/manderson/vlm4rs/fmow/fmow-rgb/train'

gpt_params = 'overlap3-1024'

csv_out = '/home/gridsan/manderson/vlm4rs/fmow/fmow_mm_data.csv'
```

```python
meta_csv = '/home/gridsan/manderson/vlm4rs/fmow/fmow-meta/train-latest.csv'
```

```python
meta_df = pd.read_csv(meta_csv)
```

```python
meta_df.head()
```

```python
with open(csv_out, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)  # Use csv.writer instead of csv.DictWriter

    writer.writerow(['filepath', 'caption'])
    
    for i in tqdm(range(len(meta_df))):
        row = meta_df.iloc[i]
        img_id = row['img_id']
        cat_id = ('_').join(img_id.split('_')[:-2])
        short_id = ('_').join(img_id.split('_')[:-1])
        txt_file = f'{txt_dir}/{cat_id}/{short_id}_gpt-{gpt_params}.txt'
        rgb_file = f'{img_dir}/{cat_id}/{short_id}/{img_id}_rgb.jpg'

        with open(txt_file, "r", encoding="utf-8") as file:
            caption = file.read()

        row = [rgb_file, caption]

        writer.writerow(row)
```
